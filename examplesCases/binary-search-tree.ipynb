```python
from ejahdilandatastruct.tree import BinarySearchTree
```

```python
# Create Binary Search Tree object
bst = BinarySearchTree()

print("Initial BinarySearchTree object:", bst)

# Test insert
bst.insert(8)
bst.insert(3)
bst.insert(10)
bst.insert(1)
bst.insert(6)
bst.insert(14)
bst.insert(4)
bst.insert(7)
bst.insert(13)

print("After insertions:", bst)

expected_insert = "1 -> 3 -> 4 -> 6 -> 7 -> 8 -> 10 -> 13 -> 14"
print("Test Insert:", "Passed" if str(bst) == expected_insert else f"Failed (Got {bst})")

# Test search
print("Searching for 6:", bst.search(6))
print("Test Search 6:", "Passed" if bst.search(6) else "Failed (Expected True)")

print("Searching for 15:", bst.search(15))
print("Test Search 15:", "Passed" if not bst.search(15) else "Failed (Expected False)")

# Test inorder traversal
inorder_result = bst.inorder_traversal()
print("Inorder Traversal:", inorder_result)

expected_traversal = [1, 3, 4, 6, 7, 8, 10, 13, 14]
print("Test Traversal:", "Passed" if inorder_result == expected_traversal else f"Failed (Got {inorder_result})")
```
