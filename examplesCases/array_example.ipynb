```python
from ejahdilandatastruct.arrays import Array
```

```python
arr = Array()

# Insert elements
arr.insert(0, 5)
arr.insert(1, 10)
arr.insert(2, 15)
arr.insert(1, 7)

# Show and test array contents
print("Current array after inserts:", arr)
expected = "Array: [5, 7, 10, 15]"
print("Test Insert:", "Passed" if str(arr) == expected else f"Failed (Got {arr})")

# Test search
print("Searching for 10, got index:", arr.search(10))
print("Test Search 10:", "Passed" if arr.search(10) == 2 else f"Failed (Got {arr.search(10)})")

print("Searching for 20, got index:", arr.search(20))
print("Test Search 20:", "Passed" if arr.search(20) == -1 else f"Failed (Got {arr.search(20)})")

# Test deletion
arr.delete(2)
print("Current array after deleting index 2:", arr)
expected_after_deletion = "Array: [5, 7, 15]"
print("Test Delete:", "Passed" if str(arr) == expected_after_deletion else f"Failed (Got {arr})")

# Test traversal
print("Traverse output:", arr.traverse())
expected_traversal = [5, 7, 15]
print("Test Traverse:", "Passed" if arr.traverse() == expected_traversal else f"Failed (Got {arr.traverse()})")

# Test length and is_empty
print("Current length:", arr.length())
print("Test Length:", "Passed" if arr.length() == 3 else f"Failed (Got {arr.length()})")

print("Is array empty before clearing?:", arr.is_empty())
print("Test Is Empty (before clearing):", "Passed" if not arr.is_empty() else "Failed (Expected not empty)")

# Clear the array
arr.delete(0)
arr.delete(0)
arr.delete(0)
print("Array after clearing:", arr)
print("Test After Clearing:", "Passed" if str(arr) == "Array: []" else f"Failed (Got {arr})")

print("Is array empty now?:", arr.is_empty())
print("Test Is Empty (after clearing):", "Passed" if arr.is_empty() else "Failed (Expected empty)")
```
