```python
from ejahdilandatastruct.stack import Stack
```

```python
# Create Stack object
s = Stack()

print("Initial Stack object:", s)

# Test push
s.push(7)
s.push(14)
s.push(21)
print("After pushes:", s)

expected_push = "Stack([7, 14, 21])"
print("Test Push:", "Passed" if str(s) == expected_push else f"Failed (Got {s})")

# Test peek
if not s.is_empty():
    top = s.peek()
    print("Peek item:", top)
    print("Test Peek:", "Passed" if top == 21 else f"Failed (Got {top})")
else:
    print("Test Peek: Failed with error - Empty stack")

# Test pop
if not s.is_empty():
    removed = s.pop()
    print("Popped item:", removed)
    print("After pop:", s)
    expected_after_pop = "Stack([7, 14])"
    print("Test Pop:", "Passed" if removed == 21 and str(s) == expected_after_pop else f"Failed (Got {s})")
else:
    print("Test Pop: Failed with error - Empty stack")

# Test is_empty
print("Test is_empty (should be False):", "Passed" if not s.is_empty() else "Failed")

# Pop all to test empty condition
s.pop()
s.pop()
print("Stack after removing all elements:", s)
print("Test is_empty (should be True):", "Passed" if s.is_empty() else "Failed")

# Test pop from empty
if s.is_empty():
    print("Test Pop from empty: Passed")
else:
    print("Test Pop from empty: Failed (Expected error)")

# Test peek from empty
if s.is_empty():
    print("Test Peek from empty: Passed")
else:
    print("Test Peek from empty: Failed (Expected error)")
```
