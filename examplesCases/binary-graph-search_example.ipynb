```python
from ejahdilandatastruct.graph import BinaryGraphSearch
```

```python
# Initial sorted list
bgs = BinaryGraphSearch([3, 8, 12])

print("Initial BinaryGraphSearch object:", bgs)

# Test insert
bgs.insert(5)
bgs.insert(10)
bgs.insert(1)
bgs.insert(20)
print("After insertions:", bgs)

expected_insert = "Sorted List: [1, 3, 5, 8, 10, 12, 20]"
print("Test Insert:", "Passed" if str(bgs) == expected_insert else f"Failed (Got {bgs})")

# Test search
print("Searching for 10, got index:", bgs.search(10))
print("Test Search 10:", "Passed" if bgs.search(10) == 4 else f"Failed (Got {bgs.search(10)})")

print("Searching for 7, got index:", bgs.search(7))
print("Test Search 7:", "Passed" if bgs.search(7) == -1 else f"Failed (Got {bgs.search(7)})")

# Test traversal (in-order)
print("Inorder Traversal:", bgs.inorder_traversal())
expected_traversal = [1, 3, 5, 8, 10, 12, 20]
print("Test Traversal:", "Passed" if bgs.inorder_traversal() == expected_traversal else f"Failed (Got {bgs.inorder_traversal()})")
```
