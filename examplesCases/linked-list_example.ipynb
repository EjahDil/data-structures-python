```python
from ejahdilandatastruct.linked_lists import SinglyLinkedList
```

```python
# Initial SinglyLinkedList object
sll = SinglyLinkedList()

print("Initial SinglyLinkedList object:", sll)

# Test insert
sll.insert(10)
sll.insert(20)
sll.insert(30)
sll.insert(40)
print("After insertions:", sll)

expected_insert = "10 -> 20 -> 30 -> 40"
print("Test Insert:", "Passed" if str(sll) == expected_insert else f"Failed (Got {sll})")

# Test search
print("Searching for 20, got:", sll.search(20))
print("Test Search 20:", "Passed" if sll.search(20) else f"Failed (Got {sll.search(20)})")

print("Searching for 50, got:", sll.search(50))
print("Test Search 50:", "Passed" if not sll.search(50) else f"Failed (Got {sll.search(50)})")

# Test delete
print("Deleting 20:", "Passed" if sll.delete(20) else "Failed")
print("After deleting 20:", sll)

expected_after_delete = "10 -> 30 -> 40"
print("Test Delete 20:", "Passed" if str(sll) == expected_after_delete else f"Failed (Got {sll})")

# Test delete a non-existent item
print("Deleting 50 (non-existent):", "Passed" if not sll.delete(50) else "Failed")
print("After trying to delete 50:", sll)

# Test traverse
print("Traverse (All elements):", sll.traverse())
expected_traverse = [10, 30, 40]
print("Test Traverse:", "Passed" if sll.traverse() == expected_traverse else f"Failed (Got {sll.traverse()})")
```
