```python
from ejahdilandatastruct.queue import Queue
```

```python
# Create Queue object
q = Queue()

# Print Initial state
print("Initial Queue object:", q)

# Test enqueue
q.enqueue(5)
q.enqueue(15)
q.enqueue(25)
print("After enqueues:", q)

expected_enqueue = "Queue([5, 15, 25])"
print("Test Enqueue:", "Passed" if str(q) == expected_enqueue else f"Failed (Got {q})")

# Test peek
if not q.is_empty():
    front = q.peek()
    print("Peek item:", front)
    print("Test Peek:", "Passed" if front == 5 else f"Failed (Got {front})")
else:
    print("Test Peek: Failed with error - Empty queue")

# Test dequeue
if not q.is_empty():
    removed = q.dequeue()
    print("Dequeued item:", removed)
    print("After dequeue:", q)
    expected_after_dequeue = "Queue([15, 25])"
    print("Test Dequeue:", "Passed" if removed == 5 and str(q) == expected_after_dequeue else f"Failed (Got {q})")
else:
    print("Test Dequeue: Failed with error - Empty queue")

# Test is_empty
print("Test is_empty (should be False):", "Passed" if not q.is_empty() else "Failed")

# Dequeue all to test empty condition
q.dequeue()
q.dequeue()
print("Queue after removing all elements:", q)
print("Test is_empty (should be True):", "Passed" if q.is_empty() else "Failed")

# Test dequeue from empty
if q.is_empty():
    print("Test Dequeue from empty: Passed")
else:
    print("Test Dequeue from empty: Failed (Expected error)")

# Test peek from empty
if q.is_empty():
    print("Test Peek from empty: Passed")
else:
    print("Test Peek from empty: Failed (Expected error)")
```
